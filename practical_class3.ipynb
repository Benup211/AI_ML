```python
plt.boxplot(train_data)
plt.xticks(range(1, len(train_data.columns) + 1), train_data.columns)
plt.xlabel('Variables')
plt.ylabel('Values')
plt.title('Box Plot of Data')
plt.show()
```

```python
plt.boxplot(train_data['Size'], showfliers=False)
plt.xlabel('Size')
plt.ylabel('Values')
plt.title('Box Plot of Data (without outliers)')
plt.show()
```

```python
for column_name in ['Weight', 'Size', 'Sweetness', 'Ripeness', 'Acidity']:
# Assuming df is your DataFrame and 'column_name' is the column for which you want to check outliers
    Q1 = train_data[column_name].quantile(0.25)
    Q3 = train_data[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # Define the boundary for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Get the number of outliers
    outliers = train_data[(train_data[column_name] < lower_bound) |
                    (train_data[column_name] > upper_bound)]
    num_outliers = outliers.shape[0]

    print(f'There are {num_outliers} outliers in {column_name}')
    #replace with nan for outliers
    train_data[f'Adjusted_{column_name}'] = np.where((train_data[column_name] < lower_bound) | (
    train_data[column_name] > upper_bound), np.nan, train_data[column_name])
    # Calculate the mean of the non-outlier values
    meidan = train_data[f'Adjusted_{column_name}'].median()

# Replace the NaNs in the 'Adjusted' column with the mean
    train_data.fillna({f'Adjusted_{column_name}':meidan}, inplace=True)
    train_data = train_data.drop(columns=column_name, axis = 1)
```

```python
features=['Adjusted_Weight', 'Adjusted_Size', 'Adjusted_Sweetness', 'Adjusted_Ripeness', 'Adjusted_Acidity']
label=['Quality']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_data[features], train_data[label], test_size=0.3, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = logistic_model.predict(X_test)
accuracy_score(y_test, y_pred)*100
```

```python
X_test_list=X_train.iloc[0]
X_test_list
```

```python
predicted_model=logistic_model.predict([X_test_list])
label_encoder.inverse_transform(predicted_model)
```

```python
import pickle

with open('banana_quality_model.pkl', 'wb') as file:
    pickle.dump(logistic_model, file)
with open('banana_quality_label_encoder.pkl', 'wb') as file:
    pickle.dump(label_encoder, file)
```
