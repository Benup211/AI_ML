```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd 
train_data=pd.read_excel('./crop.xlsx')
train_data.head()
```

```python
train_data.tail(12)
```

```python
train_data.describe()
```

```python
nan_count=train_data.isnull().sum(axis=0)
nan_count
cp_train_data=train_data.copy()
train_data=train_data.dropna(axis=0)
train_data.tail(15)
```

```python
plt.scatter(train_data['Rain Fall (mm)'],train_data['Yeild (Q/acre)'])
plt.title('Rain Fall vs Yield')
plt.xlabel('Rain Fall')
plt.ylabel('Yield')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(cp_train_data.isnull(),cbar=False,cmap='binary')
plt.show()
```

```python
corr_matrix=train_data.corr()
corr_matrix['Yeild (Q/acre)'].sort_values(ascending=False)
```

```python
plt.boxplot(train_data['Yeild (Q/acre)'])
plt.title('Box Plot for Yield')
plt.ylabel('Yield')
plt.show()
```
