```python
import pandas as pd
import numpy as np
```

```python
train_data=pd.read_csv('HepatitisCdata.csv',index_col=0)
train_data.head()
```

```python
train_data.tail()
```

```python
train_data.describe()
```

```python
train_data.isnull().sum()
```

```python
train_data.dropna(inplace=True)
train_data.reset_index(drop=True, inplace=True)
train_data.isnull().sum()
```

```python
train_data.head()
```

```python
train_data.tail()
```

```python
train_data['Sex'] = train_data['Sex'].replace({'m': 0, 'f': 1})
train_data.tail()
```

```python
train_data.Category.unique().tolist()
```

```python
train_data['Category']=train_data['Category'].map({'0=Blood Donor':0,'0s=suspect Blood Donor':1,'1=Hepatitis':2,'2=Fibrosis':2,'3=Cirrhosis':2})
train_data.head()
```

```python
train_data.Category.value_counts().plot(kind='bar')
```

```python
train_data.columns.to_list()
```

```python
features=['Age',
 'Sex',
 'ALB',
 'ALP',
 'ALT',
 'AST',
 'BIL',
 'CHE',
 'CHOL',
 'CREA',
 'GGT',
 'PROT']
Label=['Category']

from sklearn.model_selection import StratifiedShuffleSplit

sss = StratifiedShuffleSplit(n_splits=1, test_size=0.1, random_state=42)
X,y=train_data[features],train_data[Label]
for train_index, test_index in sss.split(X,y):
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

y_pred=model.predict(X_test)
accuracy_score(y_test,y_pred)*100
```

```python
cm=confusion_matrix(y_test,y_pred)
plt.figure(figsize=(10,8))
sns.heatmap(cm,annot=True,fmt='d',cmap='coolwarm')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
